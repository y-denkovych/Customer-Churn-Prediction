# Customer Churn Prediction

```python
import pandas as pd
from password import password
from sqlalchemy import create_engine
```

```python
df_1 = pd.read_csv("data/sampleSubmission.csv")
df_2 = pd.read_csv("data/test.csv")
df_3 = pd.read_csv("data/train.csv")
m1_df = df_2.merge(df_1)
main_df = m1_df.append(df_3)
main_df.drop(["id"], inplace = True, axis = 1)
main_df.head(15)
```

```python
main_df.isnull().any()
```

```python
main_df.dtypes
```

```python
main_df['churn'] = main_df['churn'].map({'yes':1 ,'no':0})
main_df.head(10)
```

```python
main_df.drop(['state', 'area_code'], inplace=True, axis=1)
main_df.head(10)
```

```python
main_df.nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
list_m = main_df.dtypes[main_df.dtypes == "object"].index.tolist()

ml_df = pd.get_dummies(main_df, prefix=list_m, drop_first=True)
ml_df
```

```python
ml_df.nunique()
```

```python
pd.set_option('display.max_rows', 100)
ml_df.isnull().any()
```

```python
ml_df.dtypes
```

```python
main_df.to_csv("data/Curated_Data.csv")
```

```python
from selenium import webdriver
import chromedriver_binary
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
chrome_options = Options()
#chrome_options.add_argument("--disable-extensions")
#chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--headless")
#chrome_options.headless = True
#chrome_options.add_argument("--start-maximized")
webdriver = webdriver.Chrome(options=chrome_options)
webdriver.implicitly_wait(10)
url = 'https://2021.ai/offerings/asset-store/customer-churn-prediction/'
webdriver.get(url)
#accept_cookies_settings = webdriver.find_element(by=By.XPATH, value='//*[@id="coiPage-1"]/div[2]/div[1]/button[3]')
#accept_cookies_settings.click()
print("Please answer the following questions: ")
print("If no values are available, please keep blank or add a zero")
webdriver.implicitly_wait(10)
time.sleep(2)
customers = input("How many customers do you currently have? --> ")
cust_worth = input("What is the average individual customer worth? --> ")
cust_lost = input("How many customers you lost last year? --> ")
model_acc = input("What is your ML model accuracy? --> ")
ret_cus = input("What is the percentage of customers that you think you can retain? --> ")
inputElement = webdriver.find_element("id", "customers")
inputElement.clear()
inputElement.send_keys(customers)
inputElement = webdriver.find_element("id", "customerValue")
inputElement.clear()
inputElement.send_keys(cust_worth)
inputElement = webdriver.find_element("id", "customersLost")
inputElement.clear()
inputElement.send_keys(cust_lost)
expand = webdriver.find_element(by=By.XPATH, value='//*[@id="post-251035"]/div/div/div/div[4]/div[5]/div/div/h3')
expand.click()
inputElement = webdriver.find_element("id", "detectChurn")
inputElement.clear()
inputElement.send_keys(model_acc)
inputElement = webdriver.find_element("id", "retainCustomers")
inputElement.clear()
inputElement.send_keys(ret_cus)
print("-----------------------------------------------")
print("-----------------------------------------------")
print("Based on the provided values:")
cr = webdriver.find_element(By.ID, 'churnRate')
print(f'  1. The Churn rate is {cr.text}%')
rl = webdriver.find_element(By.ID, 'revenueLost')
if rl.text == "NaN":
    print("  2. No Revenue data available")
else: print(f'  2. The Revenue lost is {rl.text}')
if ((model_acc.strip() != "" and ret_cus.strip() != "") | (model_acc != 0 and ret_cus != 0)):
    rd = webdriver.find_element(By.ID, 'reduceChurn')
    print(f'  3. You can reduce the Churn rate by {rd.text}%')
    rc = webdriver.find_element(By.ID, 'recoveredCustomers')
    print(f'  4. You can retain {rc.text} customers at risk')
    rl = webdriver.find_element(By.ID, 'recoveredRevenue')
    if rl.text == "NaN":
        print ("  5. No recovered Revenue data available")
    else: print(f'  5. The recovered Revenue will be {rl.text}')
    print("Thank you!")
else: print("Thank you!")
```
